# 9. how many person are not a regular person with respect to activity try to find out those 

```python
import logging as lg 
lg.basicConfig(filename = "fitbit_logging.txt",level=lg.DEBUG,format='%(asctime)s %(message)s')
m = df[["ActivityDate","Id"]].groupby("Id").count().reset_index()
def lazy(days):
  for i in m["ActivityDate"]:
    if days <= 10:
      return"Inactive"
    elif 10 < days <= 20:
      return"Active"
    elif days > 20:
      return"Very Active"

m["Days"] = m["ActivityDate"].apply(lazy)
m
```

# 10 . who is the thired most active person in this dataset find out those in pandas and in sql both .

```python
a = pd.DataFrame(df['Id'].value_counts())
a = a.reset_index()
third_most = 15
for i in range(len(a['Id'])):
    if int(a[i:i+1]['Id'])==third_most:
        print("third most active person's ID is : ",int(a[i:i+1]['index']))
```
